**Question 1:** Write a function that returns the most popular book in each subject. Use it to find the most popular book in
    the _Mystery Fiction_ subject (15pt)

```python
import turicreate.aggregate as agg

def most_popular_title_per_subject(sframe):
  #reducing the size of the sframe to the only needed columns
  sf_popular = sframe['Title', 'Subjects','ItemCount']
  
  #splitting subjects
  sf_popular['subject_list'] = sf_popular['Subjects'].apply(lambda s: s.split(","))
  sf_popular['subject_list'] = sf_popular['subject_list'].apply(lambda l: [subject.strip().lower() for subject in l])
  sf_popular = sf_popular.stack("subject_list", new_column_name="subject") 
  sf_popular = sf_popular.remove_column('Subjects')
  
  #grouping by title and subjects and counting the number of copies of books
  g = sf_popular.groupby(['Title', 'subject'],{'sum ItemCount': agg.SUM("ItemCount")})
  g.sort('sum ItemCount', ascending=False )

  # calculating the most popular title in each subject
  g_argmax = g.groupby('subject',{'Title': agg.ARGMAX('sum ItemCount','Title')})
  
  #joining tables to include the count
  g_argmax = g_argmax.join(g, on=['subject','Title'], how="inner")
  g_argmax = g_argmax.rename({'Title': 'most popular'})
  g_argmax.sort('sum ItemCount', ascending=False ).print_rows(20)
  return g_argmax
```

```python
sf_most_popular = most_popular_title_per_subject(sf)
```

```python
print("the most popular book in the Mystery Fiction is:")
print(list(sf_most_popular[sf_most_popular['subject'] == 'mystery fiction']))
```

```python
# find top-10 fiction subjects
def top_n_fiction_subjects(sframe, n=10):
  #reducing the size of the sframe to the only needed columns
  sf_popular = sframe['BibNum', 'year', 'Subjects']
  #splitting subjects
  sf_popular['subject_list'] = sf_popular['Subjects'].apply(lambda s: s.split(","))
  sf_popular['subject_list'] = sf_popular['subject_list'].apply(lambda l: [subject.strip().lower() for subject in l])
  sf_popular = sf_popular.remove_column('Subjects')
  sf_popular = sf_popular.stack("subject_list", new_column_name="subject") 
  all_subjects = sf_popular['subject'].unique()
  fiction_subjects = [subj for subj in all_subjects if 'fiction' in subj]
  # filtering
  sf_popular = sf_popular[sf_popular.apply(lambda r: r['subject'] in fiction_subjects)]
  sf_popular = sf_popular.unique()
  # sf_popular = sf_popular[sf_popular.apply(lambda r: 'fiction' in sf_popular['subject'])]
  # sf_popular.materialize()
  # sf_popular = sf_popular['fiction' in sf_popular['subject']]
  #grouping by subject and counting the number of books per subject
  g = sf_popular.groupby('subject',{'Count': agg.COUNT()})
  sf_top_n_fiction_subjects = g.sort('Count', ascending=False )[:n]
  top_n_fiction_subjects = [r["subject"] for r in sf_top_n_fiction_subjects]
  
  print(f"\ntop n fiction subjects\n: {top_n_fiction_subjects}")
  sf_popular = sframe['BibNum', 'year', 'Subjects','ItemCount']
  #splitting subjects
  sf_popular['subject_list'] = sf_popular['Subjects'].apply(lambda s: s.split(","))
  sf_popular['subject_list'] = sf_popular['subject_list'].apply(lambda l: [subject.strip().lower() for subject in l])
  sf_popular = sf_popular.remove_column('Subjects')
  sf_popular = sf_popular.stack("subject_list", new_column_name="subject") 
  # all_subjects = sf_popular['subject'].unique()
  
  sf_popular = sf_popular[sf_popular.apply(lambda r: r['subject'] in top_n_fiction_subjects)]
  sf_top_n_fiction_subjects_by_year = sf_popular.groupby(['subject','year'],{'sum ItemCount': agg.SUM("ItemCount")})
  #joining tables to include the count
  # g_argmax = g_argmax.join(g, on=['subject','Title'], how="inner")
  # g_argmax = g_argmax.rename({'Title': 'most popular'})
  # g_argmax.sort('Count', ascending=False ).print_rows(20)
  return sf_top_n_fiction_subjects_by_year
```

**Question 3:** Write a function that gets as input a subject name. The function returns the top-20 most _popular_ authors for the input subject (15pt).



```python
def subject_most_popular_authors(sframe, subject):
  #reducing the size of the sframe to the only needed columns and rows
  sf_popular = sframe['Author', 'Subjects']

  #splitting subjects
  sf_popular['subject_list'] = sf_popular['Subjects'].apply(lambda s: s.split(","))
  sf_popular['subject_list'] = sf_popular['subject_list'].apply(lambda l: [subject.strip().lower() for subject in l])
  sf_popular = sf_popular.stack("subject_list", new_column_name="subject") 
  sf_popular = sf_popular.remove_column('Subjects')
  
  sf_popular = sf_popular[sf_popular.apply(lambda r: r['subject'] == subject.lower() and r['Author'] != "")]
  #grouping by author and counting the number of them
  # g = sf_popular.groupby(['Author', 'subject'],{'Count': agg.COUNT()})
  g = sf_popular.groupby('Author',{'Count': agg.COUNT()})
  g = g.dropna()
  top_20 = g.sort('Count', ascending=False )[:20]

  return top_20
```

```python
subject_name = "mystery fiction"
sf_top_subject_authors = subject_most_popular_authors(sframe=sf, subject=subject_name)
print(f"top-20 most popular authors in the subject of: {subject_name}")
sf_top_subject_authors.print_rows(20)
```

**Question 4:** Calculate the average and median age of bloggers in each topic by gender(5pt). 
Calculate the average and median age of bloggers according to their sign (5pt). 
Visualize the age distributions  of bloggers according to their  topic and gender using Seaborn and PlotlyExpress(15pt)

```python
 # The average and median age of bloggers in each topic by gender
def calc_avg_median_age_by_topic_and_gender(sframe):
  sf_calc = sframe["gender", "age", "topic"]
  g = sf_calc.groupby(["topic","gender"], {"average age": agg.MEAN('age')}, {"median age": agg.QUANTILE('age', 0.5)})
  return g

# The average and median age of bloggers according to their sign
def calc_avg_median_age_by_sign(sframe):
  sf_calc = sframe["sign", "age"]
  g = sf_calc.groupby("sign", {"average age": agg.MEAN('age')}, {"median age": agg.QUANTILE('age', 0.5)})
  return g
```

```python
sf_avg_median_age_by_topic_and_gender = calc_avg_median_age_by_topic_and_gender(sframe=sf_blogs)
print("The average and median age of bloggers in each topic by gender")
sf_avg_median_age_by_topic_and_gender.print_rows(10)
sf_avg_median_age_by_sign = calc_avg_median_age_by_sign(sframe=sf_blogs)
print("The average and median age of bloggers according to their sign")
sf_avg_median_age_by_sign.print_rows(12)
```
